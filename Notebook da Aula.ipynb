## 4. Trabalhando DateTime e Melt

### 4.1.1 Datas com Datetime

```python
import datetime

datetime.datetime(2020,11,3)
```

```python
from datetime import date

date.today()
```

```python
meses = {
    'Jan': 1,
    'Fev': 2,
    'Mar': 3,
    'Abr': 4,
    'Mai': 5,
    'Jun': 6,
    'Jul': 7,
    'Ago': 8,
    'Set': 9,
    'Out': 10,
    'Nov': 11,
    'Dez': 12
}
def para_dia(mes):
    ano: int = int(mes[:4])
    mes: str = mes[5:]
    return date(ano, meses[mes], 1)

mensal.index = mensal.index.map(para_dia)
    
```

```python
mensal.head()
```

```python
plt.figure(figsize = (10,6))
ax = sns.lineplot(data = mensal, x = mensal.index, y = 'São Paulo')
plt.xticks(rotation=30)
plt.ylim(0,350)
plt.grid(True, linestyle = '--')
plt.show
```

### 4.2.1 Manipulação de dados com Melt

```python
#O melt serve para passar as colunas de estado para uma unica linha, e compilar tudo em um DF.
mensal_aberto = mensal.reset_index().melt(id_vars=['index'], value_vars = mensal.columns)
mensal_aberto.columns = ['dia_mes_ano','uf','gasto']
mensal_aberto['dia_mes_ano'] = mensal_aberto['dia_mes_ano'].astype('datetime64')
mensal_aberto.head()
```

```python
plt.figure(figsize = (10,6))
ax = sns.lineplot(data = mensal_aberto, x = 'dia_mes_ano', y = 'gasto', hue = 'uf')
plt.xticks(rotation=30)
plt.ylim(0,350)
plt.grid(True, linestyle = '--')
plt.show
```

```python
plt.figure(figsize = (10,6))
ax = sns.lineplot(data = mensal_aberto, x = 'mes', y = 'gasto', hue = 'ano')
plt.xticks(rotation=30)
plt.ylim(0,350)
plt.grid(True, linestyle = '--')
plt.show
```

```python
dias_por_mes = {
    1 : 31,
    2 : 28,
    3 : 31,
    4 : 30,
    5 : 31,
    6 : 30,
    7 : 31,
    8 : 31,
    9 : 30,
    10 : 31,
    11 : 30,
    12 : 31
}
```

```python
mensal_aberto['mes'].map(dias_por_mes)
```

```python
mensal_aberto['diario'] = mensal_aberto['gasto'] / mensal_aberto['mes'].map(dias_por_mes)
```

```python
plt.figure(figsize = (10,6))
ax = sns.lineplot(data = mensal_aberto.query('uf=="São Paulo"'), x = 'mes', y = 'diario', hue = 'ano')
plt.xticks(rotation=30)
plt.ylim(0,12)
plt.grid(True, linestyle = '--')
plt.show
```
